```python
import numpy as np
import pandas as pd

train = pd.read_csv('/kaggle/input/wi-ds-dl-competition-1/train.csv')
test = pd.read_csv('/kaggle/input/wi-ds-dl-competition-1/test.csv')
submission = pd.read_csv('/kaggle/input/wi-ds-dl-competition-1/sample_submission.csv')
```

```python
test_769 = train.iloc[41997:, 1:].values
x = train.iloc[:41997,1:].values
print(x.shape)
y2 = np.sum(test_769**2, axis=1).reshape(1, -1)
x2 = np.sum(x**2, axis=1).reshape(-1, 1)
xy = np.matmul(x, test_769.T)
dist = np.sqrt(abs(x2 - 2*xy + y2)) 
idx = dist.argmin(axis=0)
train.iloc[idx]
```

```python
x_tr = train.iloc[:,1:].values
x_te = test.values
y_te = train.iloc[:,0]

l = x_te.shape[0]
batch_size = 128

preds = np.empty(l, dtype=y_te.dtype)
tr_sq = np.sum(x_tr**2, axis=1).reshape(-1, 1)
```

```python
for i in range(0, l, batch_size):
    x_te_i = x_te[i:i+batch_size]
    te_sq = np.sum(x_te_i**2, axis=1).reshape(1, -1)
    cross = np.matmul(x_tr, x_te_i.T)
    dist = np.sqrt(tr_sq - 2*cross + te_sq) 
    nn_idx = dist.argmin(axis=0)
    preds[i:i+batch_size] = y_te[nn_idx]
```

```python
submission['Label'] = preds
submission
```

```python
submission.to_csv('submission.csv', index=False)
```
