```python
import numpy as np
import pandas as pd

train = pd.read_csv('/kaggle/input/wi-ds-dl-competition-1/train.csv')
test = pd.read_csv('/kaggle/input/wi-ds-dl-competition-1/test.csv')
submission = pd.read_csv('/kaggle/input/wi-ds-dl-competition-1/sample_submission.csv')
```

```python
x_tr = train.iloc[:,1:]
x_te = test.values
y_tr = train.iloc[:,0]

# For Offline Evaluation
# Held out last 12000 samples for evaluation
# x_tr = train.iloc[:30000,1:]
# y_tr = train.iloc[:30000,0]
# x_te = train.iloc[30000:,1:]
# y_te = train.iloc[30000:,0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
import time

knn = KNeighborsClassifier(n_neighbors=4, algorithm='brute', metric='euclidean', weights='distance', n_jobs=-1)
t0 = time.time()
knn.fit(x_tr, y_tr)
y_pred = knn.predict(x_te)
print("Time elapsed:", time.time()-t0, "s")
```
